```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv("Alojamiento.csv", encoding='latin-1')
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
X = data.drop(['valor_casa_mediana'], axis=1)
Y = data['valor_casa_mediana']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
train_data = X_train.join(Y_train)
```

```python
train_data
```

```python
train_data.hist(figsize=(15,8))
```

```python
train_data = train_data.join(pd.get_dummies(train_data.proximidad_oceánica)).drop(['proximidad_oceánica'], axis=1)
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True)
```

```python
train_data['hogares'] = np.log(train_data['hogares'] +1)
train_data['total_habitaciones'] = np.log(train_data['total_habitaciones'] +1)
train_data['total_dormitorios'] = np.log(train_data['total_dormitorios'] +1)
train_data['población'] = np.log(train_data['población'] +1)
```

```python
train_data.hist(figsize=(15,8))
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(x='Latitud', y='Longitud', data=train_data, hue="valor_casa_mediana")
```

```python
train_data['racion_dormitorios'] = train_data['total_dormitorios'] / train_data['total_habitaciones']
train_data['habitaciones_por_hogar'] = train_data['total_habitaciones'] / train_data['hogares']
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(), annot=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train, Y_train = train_data.drop(['valor_casa_mediana'], axis=1), train_data['valor_casa_mediana']
X_train_s = scaler.fit_transform(X_train)

reg = LinearRegression()

reg.fit(X_train_s, Y_train)
```

```python
test_data = X_test.join(Y_test)

test_data['hogares'] = np.log(test_data['hogares'] +1)
test_data['total_habitaciones'] = np.log(test_data['total_habitaciones'] +1)
test_data['total_dormitorios'] = np.log(test_data['total_dormitorios'] +1)
test_data['población'] = np.log(test_data['población'] +1)

test_data = test_data.join(pd.get_dummies(test_data.proximidad_oceánica)).drop(['proximidad_oceánica'], axis=1)

test_data['racion_dormitorios'] = test_data['total_dormitorios'] / test_data['total_habitaciones']
test_data['habitaciones_por_hogar'] = test_data['total_habitaciones'] / test_data['hogares']
```

```python
X_test, Y_test = test_data.drop(['valor_casa_mediana'], axis=1), test_data['valor_casa_mediana']
```

```python
X_test_s = scaler.transform(X_test)
```

```python
reg.score(X_test_s, Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor


forest = RandomForestRegressor()

forest.fit(X_train, Y_train)
```

```python
forest.score(X_test, Y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    'n_estimators': [3, 10, 30],
    'min_samples_split': [2, 4],
    'max_depth': [None, 4, 8]
}
grid_search = GridSearchCV(forest, param_grid, cv=5,
                           scoring='neg_mean_squared_error',
                           return_train_score=True)
grid_search.fit(X_train, Y_train)
```

```python
best_forest = grid_search.best_estimator_
```

```python
best_forest.score(X_test, Y_test)
```

```python
grid_search.best_estimator_.score(X_test, Y_test)
```
